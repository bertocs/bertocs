### MODELO 1 Random Forest

Probaré a tirar un modelo RandomForest entrenado con los datos de traininglimpio y la variable objetivo del dataset traininglabels. Con los datos de testlimpio haré predicciones usando. predict() y generaré la y_pred que es lo que busco.

### Modelo 4 Random Forest con Validación cruzada

Voy a probar el primer random forest que hice pero haciendole validación cruzada repetida con 5 folds.

```python
# Definir el modelo con los hiperparámetros actuales o los que quieras probar
modelo_rf = RandomForestClassifier(n_estimators=100, max_depth=10)
# Definir el esquema de validación cruzada (por ejemplo, 5-fold cross-validation)
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
# Listas para almacenar las predicciones y las etiquetas reales
predicciones = []
etiquetas_reales = []
# Realizar la validación cruzada y obtener las predicciones
for train_index, test_index in kf.split(traininglimpio, traininglabels):
    X_train, X_test = traininglimpio.iloc[train_index], traininglimpio.iloc[test_index]
    y_train, y_test = traininglabels.iloc[train_index], traininglabels.iloc[test_index]
    modelo_rf.fit(X_train, y_train)
    predicciones_fold = modelo_rf.predict(X_test)
    predicciones.extend(predicciones_fold)
    etiquetas_reales.extend(y_test)
# Calcular las métricas
accuracy = accuracy_score(etiquetas_reales, predicciones)
precision = precision_score(etiquetas_reales, predicciones, average='weighted')
recall = recall_score(etiquetas_reales, predicciones, average='weighted')
f1 = f1_score(etiquetas_reales, predicciones, average='weighted')
conf_matrix = confusion_matrix(etiquetas_reales, predicciones)
# Imprimir métricas
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("Confusion Matrix:\n", conf_matrix)
```

Explicación de las métricas: El modelo tiene mucha exactitud (accuracy) predice muchos positivos respecto al total de predicciones. Es muy preciso con lo que acierta muchos verdaderos positivos respecto al total de predicciones. Mucho Recall que es el acierto de verdaderos positivos respecto a los positivos reales y mucha sensibilidad que son los verdaderos positivos respecto al total de positivos del modelo. Por otro lado el F1 que es la evaluación del rendimiento del modelo que en este caso también es muy alta. Las métricas generadas son inferiores a las del primer modelo de Random Forest.

### Modelo 5 Random Forest con mejores hiperparámetros

Voy a usar el modelo previo de Random Forest con Validación cruzada repetida(estratificada) y buscaré los mejores hiperparámetros para generar un mejor modelo al anterior con grid search. De prueba los parametros que definiré son: n_estimators(nº de árboles aleatorios),max_depth(profundidad máxima de cada arbol), min_samples_split(nº de muestras de cada nodo).

```python
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)
# Definir el modelo
modelo_rf = RandomForestClassifier()
# Definir los hiperparámetros a sintonizar y sus rangos
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
# Configurar la validación cruzada repetida
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=42)
# Ejecutar la búsqueda en cuadrícula
grid_search = GridSearchCV(estimator=modelo_rf, param_grid=param_grid, cv=cv, scoring='accuracy')
grid_result = grid_search.fit(X, y)
# Imprimir los resultados
print("Mejor precisión:", grid_result.best_score_)
print("Mejores hiperparámetros:", grid_result.best_params_)
```

El grid search me ha generado que los mejores parámetros son estos, casualmente en el modelo previo de validación cruzada repetida usé el mismo max_depth y nº de estimadores. Prefiero no ejecutarlo porque pienso que el resultado será muy parecido al previo.
